```python
uniqueFuelType = df['fuelType'].unique()
```

```python
from deep_translator import GoogleTranslator
def translateClass(uniquevalues):
    translator = GoogleTranslator(target='en')
    rename = dict()
    for x in uniquevalues:
        rename[x] = translator.translate(x)
    print(rename)
    return rename
```

```python
replaceFuelType = translateClass(uniquevalues=uniqueFuelType)
```

```python
df['fuelType'].replace(replaceFuelType, inplace=True)
# df['fuelType'].replace({'Others':'other'}, inplace=True)
```
