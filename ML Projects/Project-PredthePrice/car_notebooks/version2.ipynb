```python
from deep_translator import GoogleTranslator
def translateClass(uniquevalues):
    translator = GoogleTranslator(target='en')
    rename = dict()
    for x in uniquevalues:
        rename[x] = translator.translate(x)
    print(rename)
    return rename
```

```python
rename = translateClass(vehicleTypeUnique)
```

```python
cldf['vehicleType'] = cldf['vehicleType'].replace(rename)
```

```python
cldf['vehicleType'].value_counts()
```

```python
cldf['vehicleType'] = cldf['vehicleType'].str.capitalize()
```
