# Handling Missing value

```python
data.head()
```

```python
data.shape
```

```python
outlier_indices = data[data['age'] > 100].index
outlier_indices
data.loc[outlier_indices].head()

```

```python
outlier_indices = data[data['age'] > 100].index
data = data.drop(outlier_indices)
print("Shape of DataFrame after dropping age outliers:", data.shape)
```

```python
#Handling missing values with mean and mode
data['social_class']=data['social_class'].fillna(data['social_class'].mode()[0])
data['city']=data['city'].fillna(data['city'].mode()[0])
data['primary_business']=data['primary_business'].fillna(data['primary_business'].mode()[0])
# data['secondary_business']=data['secondary_business'].fillna(data['secondary_business'].mode()[0])
data['type_of_house']=data['type_of_house'].fillna(data['type_of_house'].mode()[0])
data['sanitary_availability']=data['sanitary_availability'].fillna(data['sanitary_availability'].mode()[0])
data['water_availabity']=data['water_availabity'].fillna(data['water_availabity'].mode()[0])
data['loan_purpose']=data['loan_purpose'].fillna(data['loan_purpose'].mode()[0])
data['monthly_expenses']=data['monthly_expenses'].fillna(data['monthly_expenses'].mean())
data['home_ownership']=data['home_ownership'].fillna(data['home_ownership'].mode()[0])
```

```python
data.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder

# Instantiate the LabelEncoder
le = LabelEncoder()

# Apply label encoding to each specified column
data['secondary_business_encoded'] = le.fit_transform(data['secondary_business'])
data['type_of_house_encoded'] = le.fit_transform(data['type_of_house'])
data['sex_encoded'] = le.fit_transform(data['sex'])
data['social_class_encoded'] = le.fit_transform(data['social_class'])
data['primary_business_encoded'] = le.fit_transform(data['primary_business'])

data = data.drop(['secondary_business', 'type_of_house', 'sex', 'social_class', 'primary_business'], axis=1)
```

```python
# Create a new column for the ratio of 'annual_income' to 'monthly_expenses'
data['income_to_expenses_ratio'] = data['annual_income'] / data['monthly_expenses']

# Display the data with the new ratio column
data[['annual_income', 'monthly_expenses', 'income_to_expenses_ratio']].head()
```

```python
from sklearn.preprocessing import StandardScaler

# Columns to be scaled
columns_to_scale = ['age', 'annual_income', 'monthly_expenses',
                    'occupants_count', 'house_area',
                    'loan_tenure', 'loan_installments',
                    'income_to_expenses_ratio']

# Instantiate the StandardScaler
scaler = StandardScaler()

# Apply standard scaling to the specified columns
data[columns_to_scale] = scaler.fit_transform(data[columns_to_scale])

# Display the scaled data
scaled_data = data[columns_to_scale]
scaled_data.head()
```

```python
data.columns
```

```python
# Drop the original unscaled columns
data = data.drop(columns=columns_to_scale)

# Concatenate the scaled columns with the original data
data = pd.concat([data, scaled_data], axis=1)

# Display the merged data
data.columns
```
